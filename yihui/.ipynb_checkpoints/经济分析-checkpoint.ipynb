```python
import pandas as pd
```

```python
data=pd.read_excel('data/A1-01.xls')
# io = pd.io.excel.ExcelFile('data/A1-01.xls')

# data=pd.read_excel(io,sheet_name='Sheet2')
```

```python
data=data.dropna()
```

```python
data=data.drop(index=1)
```

```python
data.sort_values('法人',ascending=False)
```

```python
area=data.sort_values('法人',ascending=True)['地区']
count=data.sort_values('法人',ascending=True)['法人']
```

```python
type(data)
```

```python
import plotly.offline as py
from plotly.graph_objs import Scatter, Layout
import plotly.graph_objs as go
py.init_notebook_mode(connected=True)


data = [go.Bar(
            x=count,
            y=area,
            orientation = 'h'
)]
layout = go.Layout(
            title = '全国法人数量'
    )
figure = go.Figure(data = data, layout = layout)
py.iplot(figure)
```

```python
import plotly.offline as py
from plotly.graph_objs import Scatter, Layout
import plotly.graph_objs as go

py.init_notebook_mode(connected=True)
trace_1 = go.Bar(
            x = ["西南石油", "东方明珠", "海泰发展"],
            y = [4.12, 5.32, 0.60],
            name = "201609"
    )
trace_2 = go.Bar(
            x = ["西南石油", "东方明珠", "海泰发展"],
            y = [3.65, 6.14, 0.58],
            name = "201612"
    )

trace_3 = go.Bar(
            x = ["西南石油", "东方明珠", "海泰发展"],
            y = [2.15, 1.35, 0.19],
            name = "201703"
    )
trace = [trace_1, trace_2, trace_3]
# Layout
layout = go.Layout(
            title = '净资产收益率对比图'
    )
# Figure
figure = go.Figure(data = trace, layout = layout)
# Plot
# pyplt(figure, filename='tmp/2.html')
py.iplot(figure)
```
